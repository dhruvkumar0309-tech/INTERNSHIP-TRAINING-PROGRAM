```python

#Movie Ticket Booking

movies = {
    1: {"name": "Animal", "price": 200},
    2: {"name": "KGF 2", "price": 250},
    3: {"name": "Pushpa 2", "price": 220}
}

print("------ MOVIE TICKET BOOKING ------")
print("Available Movies:")
for key, m in movies.items():
    print(f"{key}. {m['name']} - Rs {m['price']}")

choice = int(input("\nEnter Movie Number: "))

if choice in movies:
    movie = movies[choice]
    print(f"\nYou selected: {movie['name']}")
    
    tickets = int(input("How many tickets? "))
    
    total = movie['price'] * tickets
    print("\n------ BOOKING CONFIRMATION ------")
    print("Movie:", movie["name"])
    print("Tickets:", tickets)
    print("Total Amount: Rs", total)
    print("----------------------------------")
else:
    print("Invalid Movie Number!")
```
